### Model Building

```python
x = flood_data.drop(columns=['SUBDIVISION','FLOODS'],axis=1)
```

```python
y=flood_data["FLOODS"]
```

### Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
for col in x:
    x[col]=label_encoder.fit_transform(x[col].values)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
cols=['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC', ' ANNUAL RAINFALL']
```

```python
for col in cols:
  a = x[col].values.reshape(-1,1)
  x[col]=ss.fit_transform(a)
```

### Train Test data split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
models = [LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier(),DecisionTreeClassifier()]
```

### Comparing models

```python
def compare_models_cross_validation():
  for model in models:
      cv_score = cross_val_score(model,x_train,y_train,cv=5)

      mean_accuracy = sum(cv_score)/5
      mean_accuracy = mean_accuracy*100
      mean_accuracy = round(mean_accuracy,2)
      cv_precision_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='precision')
      mean_precision = sum(cv_precision_scores) / len(cv_precision_scores)
      cv_f1_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='f1')
      mean_f1_score = sum(cv_f1_scores) / len(cv_f1_scores)
      cv_recall_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='recall')
      mean_recall_score = sum(cv_recall_scores) / len(cv_recall_scores)



      print('Cross validation for ',model , '= ',cv_score)
      print('mean accuracy ',model,mean_accuracy)
      print("mean_f1_score ",model, ' = ',mean_f1_score)
      print("mean_precision_score ",model, ' = ',mean_precision)
      print("mean_recall_score ",model, ' = ',mean_recall_score)

    # print(y_pred)
```

```python
compare_models_cross_validation()
```

* Best model is RandomForestClassifier.
* So it can be used in model building

```python
best_model =RandomForestClassifier()
```

```python
best_model.fit(x_train,y_train)
```

```python
y_predict=best_model.predict(x_test)
```
